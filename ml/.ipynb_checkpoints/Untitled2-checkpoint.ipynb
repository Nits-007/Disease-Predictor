```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time

def get_top_5_doctor_info(place, speciality):
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    url = f'https://www.practo.com/search/doctors?results_type=doctor&q=%5B%7B%22word%22%3A%22{speciality}%22%2C%22autocompleted%22%3Atrue%2C%22category%22%3A%22subspeciality%22%7D%5D&city={place}'
    driver.get(url)
    time.sleep(2) 
    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CLASS_NAME, "info-section")))
    info_sections = driver.find_elements(By.CLASS_NAME, 'info-section')

    doctor_info_list = []

    for i, info_section in enumerate(info_sections[1:6]):
        doctor_info = {}
        
        try:
            practice_locality_element = info_section.find_element(By.CSS_SELECTOR, '[data-qa-id="practice_locality"]')
            doctor_info['Location'] = practice_locality_element.text
        except NoSuchElementException:
            doctor_info['Location'] = 'Not available'

        try:
            doctor_clinic_element = info_section.find_element(By.CSS_SELECTOR, '[data-qa-id="doctor_clinic_name"]')
            doctor_info["Doctor's Clinic Name"] = doctor_clinic_element.text
        except NoSuchElementException:
            doctor_info["Doctor's Clinic Name"] = 'Not available'

        try:
            consultation_fee_element = info_section.find_element(By.CSS_SELECTOR, '[data-qa-id="consultation_fee"]')
            doctor_info['Consultation Fee'] = consultation_fee_element.text
        except NoSuchElementException:
            doctor_info['Consultation Fee'] = 'Not available'

        try:
            doctor_recommendation_element = info_section.find_element(By.CSS_SELECTOR, '[data-qa-id="doctor_recommendation"]')
            doctor_info["Positive Feedback"] = doctor_recommendation_element.text
        except NoSuchElementException:
            doctor_info["Positive Feedback"] = 'Not available'

        doctor_info_list.append(doctor_info)
    
    driver.quit()
    return doctor_info_list

doctor_info_list = get_top_5_doctor_info('Allahabad', 'Dentist')
for i, doctor_info in enumerate(doctor_info_list):
    print(f"\nDoctor {i + 1} Information:")
    for key, value in doctor_info.items():
        print(f"{key}: {value}")
```
